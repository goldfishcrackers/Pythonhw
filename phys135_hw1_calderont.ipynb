```python
import numpy as np 
from scipy.stats import norm
import matplotlib.pyplot as plt

#found mean and signma through norm.fit
mean,sigma = norm.fit(arr1)
plt.hist (arr1, bins = 30, density = True, color ='#80ffff')

#scaling for the gaussian distirbution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
y = norm.pdf(x, mean, sigma)

#plotted with color choice
plt.plot(x,y, color = '#dd99ff')

plt.savefig('hw1_q4histwgauss.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np

# gaussian function 
def Gauss(x, a, mu, sigma):
    return a * np.exp(-(x - mu)**2 / (2 * sigma**2))

#curvefit
popt,pcov = curve_fit(Gauss, x, y, p0=[max(y), mean, sigma])

#save fig
plt.savefig('hw1_q5fitvsdata.png')

#personality  for plots data and theory
plt.plot(x, y, 'b+:', label='data') #theory
plt.plot(x, Gauss(x, *popt), 'r-', label='fit') #fit
plt.legend()
plt.title('gauss theor vs data')
plt.show()

print("Differences between the theoretical and the actual gaussian fit is that they can provide use with varaiances due to environment. There will often be more noise when it comes to the experimental that will not be shown with  the theoretical because the theoretical is the perfect ideal version.")
```
